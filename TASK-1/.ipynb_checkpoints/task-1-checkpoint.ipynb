### Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### loading given data 

```python
data_url = "http://bit.ly/w-data"
data = pd.read_csv(data_url)
print(data.head())
```

###  Visualizing the data from understanding the relationship between study hours and scores

```python
plt.scatter(data['Hours'], data['Scores'], color='red')
plt.title('Hours vs. Percentage')
plt.xlabel('Hours of Study')
plt.ylabel('Percentage Score')
plt.show()
```

### Now preparing the data for training

```python
X = data[['Hours']].values
y = data['Scores'].values
```

### now slpiting the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### simple linear regression model training

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

### Making predictions using the trained model

```python
y_pred = regressor.predict(X_test)
hours = 9.25
predicted_score = regressor.predict([[hours]])
```

### Displaying the predicted score

```python
print(f"Predicted score for a student who studies for {hours} hrs/day: {predicted_score[0]:.2f}")
```

### Visualizing the regression line and the data points

```python
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('Hours vs. Percentage')
plt.xlabel('Hours of Study')
plt.ylabel('Percentage Score')
plt.show()
```
