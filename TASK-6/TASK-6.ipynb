* Prediction using Decision Tree

### Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree
```

###  loading dataset directly into NumPy arrays

```python
iris = load_iris()
X = iris.data
y = iris.target
```

### Create a Decision Tree classifier

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X, y)
```

### Visualize the Decision Tree graphically

```python
fig, ax = plt.subplots(figsize=(12, 12))
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, fontsize=10)
plt.show()
```

### now feeding new data 

```python
new_data = np.array([[5.1, 3.5, 1.4, 0.2]]) 
predicted_class = clf.predict(new_data)
predicted_species = iris.target_names[predicted_class][0]
```

```python
print(f"Predicted Class: {predicted_class}")
print(f"Predicted Species: {predicted_species}")
```
